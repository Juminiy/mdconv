```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model, model_selection 

X, y = datasets.load_diabetes(return_X_y=True)
print(X.shape,X[0], sep='\n')

X_2 = X[:,2]
X_2_r = X_2.reshape((-1,1))
print(X_2_r.shape, X_2_r[0], sep='\n')
# print(X_2)

xtrain,xtest,ytrain,ytest = model_selection.train_test_split(X_2_r, y, test_size=0.33)
lmodel=linear_model.LinearRegression()
lmodel.fit(xtrain,ytrain)

predict_y=lmodel.predict(xtest)
plt.scatter(xtest,ytest, color='red')
plt.plot(xtest,predict_y,color='black',linewidth=3)
plt.xlabel('BMIs')
plt.ylabel('Disease Progression')
plt.title('BMI scaleX, DiseaseProgression scaleY')
plt.show()
```
